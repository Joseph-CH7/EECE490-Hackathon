**Advanced Business Analytics: Revenue EDA, Performance Segmenting, and Time-Series Forecasting**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
sns.set(style="whitegrid")

#DATA FILTERING
#Filtering datafile Monthly sales

raw = pd.read_csv('/content/Monthly sales.csv', header=None, encoding='utf-8-sig')

months_full = ["January","February","March","April","May","June","July","August","September"]
month_short = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep"]
full_to_short = dict(zip(months_full, month_short))

# find all January and October headers
jan_headers = raw[raw.apply(lambda r: r.astype(str).str.contains(r'\bJanuary\b', na=False).any(), axis=1)].index.tolist()
oct_headers = raw[raw.apply(lambda r: r.astype(str).str.contains(r'\bOctober\b', na=False).any(), axis=1)].index.tolist()

print("January header rows:", jan_headers)
print("October header rows:", oct_headers)

if not jan_headers:
    raise ValueError("No January header found.")
if not oct_headers:
    raise ValueError("No October header found.")

first_oct = oct_headers[0]  # everything before this is Jan–Sep sections

sections = []

for i, h in enumerate(jan_headers):
    # section boundaries
    start = h + 1
    end = jan_headers[i+1] if i+1 < len(jan_headers) else first_oct
    if start >= first_oct:
        break

    header = raw.loc[h]

    # detect which columns correspond to Jan..Sep by scanning header row
    month_cols = {}
    for col in raw.columns:
        cell = str(header[col])
        for m in months_full:
            if re.search(rf"\b{m}\b", cell, flags=re.IGNORECASE):
                month_cols[full_to_short[m]] = col

    # need at least January to be valid
    if "Jan" not in month_cols:
        continue

    block = raw.iloc[start:end].copy()

    # keep only branch rows
    block = block[block[1].astype(str).str.contains('Stories', na=False)].copy()

    # year forward fill within this block
    block[0] = block[0].replace('', np.nan).ffill()

    # build extraction columns dynamically
    cols = [0, 1] + [month_cols[m] for m in month_short if m in month_cols]
    block = block[cols].copy()

    # rename to Year, Branch, Jan..Sep (only those present)
    rename_map = {0: "Year", 1: "Branch"}
    for m in month_short:
        if m in month_cols:
            rename_map[month_cols[m]] = m
    block = block.rename(columns=rename_map)

    # numeric conversion for month columns that exist
    for m in month_short:
        if m in block.columns:
            block[m] = pd.to_numeric(block[m].astype(str).str.replace(',','',regex=False), errors='coerce')

    block["Year"] = pd.to_numeric(block["Year"], errors="coerce").ffill()
    block["Branch"] = block["Branch"].astype(str).str.strip()

    sections.append(block)

df_a = pd.concat(sections, ignore_index=True)
df_a = df_a.dropna(subset=["Year","Branch"])
df_a["Year"] = df_a["Year"].astype(int)

# keep last occurrence if duplicates exist across repeated pages
df_a = df_a.drop_duplicates(["Year","Branch"], keep="last")

# ---- Oct-Dec block
start_b = first_oct + 1
df_b_raw = raw.iloc[start_b:].copy()

df_b = df_b_raw[df_b_raw[1].astype(str).str.contains('Stories', na=False)].copy()
df_b[0] = df_b[0].replace('', np.nan).ffill()

df_b = df_b[[0,1,2,3,4,5]].copy()
df_b.columns = ['Year','Branch','Oct','Nov','Dec','Total']

for c in ['Oct','Nov','Dec','Total']:
    df_b[c] = pd.to_numeric(df_b[c].astype(str).str.replace(',','',regex=False), errors='coerce')

df_b['Year'] = pd.to_numeric(df_b['Year'], errors='coerce').ffill().astype(int)
df_b['Branch'] = df_b['Branch'].astype(str).str.strip()
df_b = df_b.drop_duplicates(['Year','Branch'], keep='first')

# ---- Merge
df_wide = df_a.merge(df_b, on=['Year','Branch'], how='left')

print("Non-zero Jan 2026 count:", (df_wide[df_wide['Year']==2026]['Jan'] > 0).sum())
display(df_wide[df_wide['Year']==2026][['Branch','Jan','Feb','Mar']].sort_values('Jan', ascending=False).head(20))



# Filtering Category summary
FILE_PATH = "/content/Category summary.csv"

# Excel column letters you want to drop (examples: D, H, J)
DROP_LETTERS = ["D", "H", "J"]


# If the number of columns doesn't match, the code will auto-name them.
PREFERRED_NAMES_7 = [
    "Category",
    "Qty",
    "Total_Price",
    "Total_Cost",
    "Cost_Percent",
    "Total_Profit",
    "Profit_Percent"
]
# ---------------------------------------------

def excel_letter_to_index(letter: str) -> int:
    """Convert Excel column letter (A, B, ..., Z, AA, AB, ...) to 0-based index."""
    letter = letter.strip().upper()
    idx = 0
    for ch in letter:
        idx = idx * 26 + (ord(ch) - ord("A") + 1)
    return idx - 1  # 0-based

def clean_report_csv(
    file_path: str,
    drop_letters=None,
    preferred_names_7=None,
):
    # 1) Read raw (no header)
    raw = pd.read_csv(file_path, header=None, encoding="utf-8-sig")

    df = raw.copy()

    # 2) Drop unwanted columns (by Excel letters)
    if drop_letters:
        drop_idx = [excel_letter_to_index(x) for x in drop_letters]
        drop_idx = [i for i in drop_idx if 0 <= i < df.shape[1]]
        if drop_idx:
            df = df.drop(df.columns[drop_idx], axis=1)

    # 3) Remove rows that don’t contain ANY numbers (removes titles/headers/page lines)
    df_num = df.apply(pd.to_numeric, errors="coerce")
    df = df[df_num.notna().any(axis=1)].copy()
    df.reset_index(drop=True, inplace=True)

    # 4) Rename columns
    # If exactly 7 cols, use preferred_names_7 if given
    if preferred_names_7 and df.shape[1] == 7 and len(preferred_names_7) == 7:
        df.columns = preferred_names_7
    else:
        df.columns = [f"col_{i}" for i in range(df.shape[1])]

    # 5) Convert numeric columns (remove commas first)
    # If there is a "Category" column, keep it as text; otherwise convert all columns that look numeric.
    if "Category" in df.columns:
        non_numeric_cols = {"Category"}
    else:
        non_numeric_cols = set()

    for c in df.columns:
        if c in non_numeric_cols:
            df[c] = df[c].astype(str).str.strip()
        else:
            df[c] = pd.to_numeric(
                df[c].astype(str).str.replace(",", "", regex=False).str.strip(),
                errors="coerce"
            )

    # 6) Drop fully empty columns (sometimes happen after filtering)
    df = df.dropna(axis=1, how="all")

    return raw, df

# ---------- RUN ----------
raw, df_clean = clean_report_csv(
    FILE_PATH,
    drop_letters=DROP_LETTERS,
    preferred_names_7=PREFERRED_NAMES_7
)

print("RAW shape:", raw.shape)
print("CLEAN shape:", df_clean.shape)
print("CLEAN columns:", list(df_clean.columns))

display(df_clean.head(15))
display(df_clean.describe(include="all").T)
df_clean.info()

#Filtering Product Profitability:
FILE_PATH = "/content/Product profitability.csv"
NEW_DF_NAME = "df_product"
DROP_COLUMNS_INDEX = [3, 7, 9]   # D, H, J
# ============================

# Read raw file (no header)
raw_product = pd.read_csv(FILE_PATH, header=None, encoding="utf-8-sig")

# 2Drop unwanted columns safely
drop_idx = [i for i in DROP_COLUMNS_INDEX if i < raw_product.shape[1]]
df_product = raw_product.drop(raw_product.columns[drop_idx], axis=1)

# Keep only rows that contain numbers
df_numeric = df_product.apply(pd.to_numeric, errors="coerce")
df_product = df_product[df_numeric.notna().any(axis=1)].copy()
df_product.reset_index(drop=True, inplace=True)

# Rename columns (adjust count automatically)
if df_product.shape[1] == 7:
    df_product.columns = [
        "Product",
        "Qty",
        "Total_Price",
        "Total_Cost",
        "Cost_Percent",
        "Total_Profit",
        "Profit_Percent"
    ]
else:
    df_product.columns = [f"col_{i}" for i in range(df_product.shape[1])]

# Convert numeric columns
for col in df_product.columns:
    if col.lower() != "product":
        df_product[col] = pd.to_numeric(
            df_product[col].astype(str).str.replace(",", "", regex=False).str.strip(),
            errors="coerce"
        )
    else:
        df_product[col] = df_product[col].astype(str).str.strip()

# Final clean preview
print("Final shape:", df_product.shape)
display(df_product.head(15))
df_product.info()

#Filtering Sales by Group
FILE_PATH = "/content/Sales by Groups.csv"

# Read raw
raw_groups = pd.read_csv(FILE_PATH, header=None, encoding="utf-8-sig")

# Drop B and E (indexes 1 and 4)
df_groups = raw_groups.drop(raw_groups.columns[[1, 4]], axis=1)

# Keep only rows that contain numbers
df_numeric = df_groups.apply(pd.to_numeric, errors="coerce")
df_groups = df_groups[df_numeric.notna().any(axis=1)].copy()
df_groups.reset_index(drop=True, inplace=True)

# Rename columns
df_groups.columns = ["Product", "Quantity", "Amount"]

# Convert numeric columns
df_groups["Quantity"] = pd.to_numeric(
    df_groups["Quantity"].astype(str).str.replace(",", "", regex=False),
    errors="coerce"
)

df_groups["Amount"] = pd.to_numeric(
    df_groups["Amount"].astype(str).str.replace(",", "", regex=False),
    errors="coerce"
)

df_groups["Product"] = df_groups["Product"].astype(str).str.strip()

# Preview
print("Final shape:", df_groups.shape)
display(df_groups.head(15))
df_groups.info()


print("ANALYSIS STARTS HERE:")

######## PART 1 ################

# ------------------------------------------------------------
# PART 1 — Exploratory Data Analysis (EDA)
# Objective:
# - Understand revenue distribution per branch
# - Identify revenue concentration
# - Analyze seasonality patterns
# ------------------------------------------------------------

# Define month columns
month_cols = ["Jan","Feb","Mar","Apr","May","Jun",
              "Jul","Aug","Sep","Oct","Nov","Dec"]

# Create total revenue column per branch per year
df_wide["Total_Revenue"] = df_wide[month_cols].sum(axis=1)

# Loop per year to analyze revenue distribution
for year in sorted(df_wide["Year"].unique()):
    print(f"\n==================== {year} ====================\n")

    # Filter dataset for the given year
    df_year = df_wide[df_wide["Year"] == year].copy()

    # Compute total company revenue for that year
    company_total = df_year["Total_Revenue"].sum()

    # Compute revenue share percentage per branch
    df_year["Revenue_%"] = (
        df_year["Total_Revenue"] / company_total * 100
    )

    # Sort branches by total revenue descending
    df_year = df_year.sort_values("Total_Revenue", ascending=False)

    # Print branch revenue contribution
    print(
        df_year[["Branch","Total_Revenue","Revenue_%"]]
        .to_string(index=False)
    )

    # Print company total
    print("\nCompany Total:", round(company_total,2))


# ------------------------------------------------------------
# Seasonality Pattern Analysis (Company Level)
# Objective:
# - Identify strongest and weakest months
# - Visualize demand cyclicality
# ------------------------------------------------------------
monthly_totals = df_wide[df_wide["Year"]==2025][
    ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep",
     "Oct","Nov","Dec"]
].sum()

monthly_totals.plot(kind="bar", figsize=(10,5))
plt.title("Total Monthly Sales - 2025")
plt.show()


##### PART 2 ##########

# ------------------------------------------------------------
# PART 2 — Forecasting Using Prophet
# ------------------------------------------------------------

# Try importing Prophet, install if not available
try:
    from prophet import Prophet
except ImportError:
    !pip -q install prophet
    from prophet import Prophet

# -----------------------------
# 0) Build CLEAN long dataframe from df_wide
# Objective:
# - Convert wide (Year × Branch × Month) format
# - Into long time-series format (Branch × Date × Sales)
# -----------------------------
months_order = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
month_map = {m:i+1 for i,m in enumerate(months_order)}

df_long = df_wide.melt(
    id_vars=["Year","Branch"],
    value_vars=[m for m in months_order if m in df_wide.columns],
    var_name="Month",
    value_name="Sales"
)

# Clean and enforce proper data types
df_long["Branch"] = df_long["Branch"].astype(str).str.strip()
df_long["Year"]   = pd.to_numeric(df_long["Year"], errors="coerce")
df_long["Sales"]  = pd.to_numeric(df_long["Sales"], errors="coerce")
df_long["Month_Num"] = df_long["Month"].map(month_map)

# Create proper datetime index (monthly frequency)
df_long["Date"] = pd.to_datetime(
    dict(year=df_long["Year"], month=df_long["Month_Num"], day=1),
    errors="coerce"
)

# Remove invalid rows and sort properly
df_long = df_long.dropna(subset=["Branch","Date","Sales"])
df_long = df_long[["Branch","Date","Sales"]].sort_values(["Branch","Date"]).reset_index(drop=True)

# -----------------------------
# 1) Trim time series
# Objective:
# - Remove leading zeros before branch opening
# - Remove trailing placeholder zeros
# -----------------------------
def trim_series(g):
    g = g.sort_values("Date").reset_index(drop=True)

    # Find non-zero sales
    nz = g[g["Sales"] > 0]
    if nz.empty:
        return g.iloc[0:0]

    # Remove leading zero period
    first_nz_date = nz["Date"].min()
    g = g[g["Date"] >= first_nz_date].copy()

    # Remove trailing placeholder zeros
    last_nz_date = g.loc[g["Sales"] > 0, "Date"].max()
    g = g[g["Date"] <= last_nz_date].copy()

    return g

df_ts = df_long.groupby("Branch", group_keys=False).apply(trim_series)

# Deduplicate in case raw export repeated report pages
df_ts = (df_ts.groupby(["Branch","Date"], as_index=False)["Sales"]
              .sum()
              .sort_values(["Branch","Date"])
              .reset_index(drop=True))

print("After trimming & dedupe:", df_ts.shape)

# -----------------------------
# 2) Keep only STABLE branches
# Criteria:
# - At least 8 months history
# - Last 3 months positive (active)
# - Total sales positive
# -----------------------------
def keep_stable_branch(g, min_months=8, tail_months=3):
    g = g.sort_values("Date")
    if len(g) < min_months:
        return False
    if (g.tail(tail_months)["Sales"] <= 0).any():
        return False
    if g["Sales"].sum() <= 0:
        return False
    return True

valid_branches = [
    b for b, g in df_ts.groupby("Branch")
    if keep_stable_branch(g, min_months=8, tail_months=3)
]

df_ts_stable = df_ts[df_ts["Branch"].isin(valid_branches)].copy()

print("Stable branches kept:", len(valid_branches))
display(pd.Series(valid_branches).head(20))

# ------------------------------------------------------------
# Prophet forecasting begins below.
# Model training is done per branch and can take time.
# Estimated runtime: ~2 to 2.5 minutes depending on machine.
# ------------------------------------------------------------
print("\n[INFO] Starting Prophet model training and forecasting per stable branch.")
print("[INFO] This step may take approximately 2–3 minutes due to model fitting.\n")

# -----------------------------
# 3) Prophet forecast Feb–Dec 2026 per stable branch
# -----------------------------
TARGET_END = pd.Timestamp("2026-12-01")
pred_rows = []

# Loop through each stable branch and train separate Prophet model
for branch, g in df_ts_stable.groupby("Branch"):
    g = g.sort_values("Date")

    if len(g) < 8:
        continue

    dfp = g.rename(columns={"Date":"ds", "Sales":"y"})[["ds","y"]]

    # Prophet configuration
    model = Prophet(
        yearly_seasonality=True,
        weekly_seasonality=False,
        daily_seasonality=False,
        seasonality_mode="additive",
        changepoint_prior_scale=0.2
    )
    model.fit(dfp)

    # Determine number of months to forecast
    last_date = dfp["ds"].max()
    months_to_forecast = (TARGET_END.to_period("M") - last_date.to_period("M")).n
    if months_to_forecast <= 0:
        continue

    # Generate future dataframe
    future = model.make_future_dataframe(periods=months_to_forecast, freq="MS", include_history=False)
    fcst = model.predict(future)[["ds","yhat"]]

    # Keep only Feb–Dec 2026 forecasts
    fcst = fcst[(fcst["ds"] >= "2026-02-01") & (fcst["ds"] <= "2026-12-01")].copy()
    fcst["yhat"] = fcst["yhat"].clip(lower=0)

    # Append forecast rows
    for _, r in fcst.iterrows():
        pred_rows.append({"Branch": branch, "Date": r["ds"], "Predicted_Sales": float(r["yhat"])})

df_pred_2026 = pd.DataFrame(pred_rows)

# Safety check
if df_pred_2026.empty:
    raise ValueError("No forecasts generated. Try lowering min_months or tail_months in keep_stable_branch().")

# Final formatting
df_pred_2026 = df_pred_2026.sort_values(["Branch","Date"]).reset_index(drop=True)
df_pred_2026["Predicted_Sales"] = df_pred_2026["Predicted_Sales"].round(0).astype(int)

print("Forecast rows (Feb–Dec 2026):", df_pred_2026.shape)
display(df_pred_2026.head(30))

# -----------------------------
# 4) Convert forecast to wide format
# -----------------------------
df_pred_2026["Month"] = df_pred_2026["Date"].dt.strftime("%b")

pred_wide_2026 = (df_pred_2026.pivot_table(index="Branch", columns="Month",
                                          values="Predicted_Sales", aggfunc="sum")
                  .reindex(columns=["Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]))

pred_wide_2026["Total_Feb_Dec_2026"] = pred_wide_2026.sum(axis=1)

print("\nTop predicted branches (Feb–Dec 2026):")
display(pred_wide_2026.sort_values("Total_Feb_Dec_2026", ascending=False).head(15))

# ------------------------------------------------------------
# Plot sections below remain unchanged.
# They visualize:
# - Company evolution
# - Top 3 branch forecasts
# - Historical + forecast comparison
# - Combined evolution of top 3 branches
# ------------------------------------------------------------
```

```python
# =========================
# FULL CLEANING PIPELINE (4 files)
# Purpose:
# - Reads 4 raw POS-export CSV files (messy format, repeated report headers, mixed text + numbers)
# - Cleans and standardizes them into 4 analysis-ready tables
# Outputs:
#   1) monthly_wide    : (Year, Branch, Jan..Dec, Total)
#   2) category_branch : (Branch, Category, Qty, Revenue, Total_Cost, Total_Profit, Margin, ...)
#   3) product_profit  : (Branch, Service, Category, Section, Product, Qty, Revenue, Total_Cost, Total_Profit, Margin, ...)
#   4) groups_sales    : (Branch, Division, Group, Product, Quantity, Amount)
# =========================

import os
import re
import numpy as np
import pandas as pd

# Candidate file locations for Colab (/content) and local sandbox (/mnt/data)
CANDIDATE_PATHS = {
    "monthly": ["/mnt/data/Monthly sales.csv", "/content/Monthly sales.csv"],
    "category": ["/mnt/data/Category summary.csv", "/content/Category summary.csv"],
    "profit": ["/mnt/data/Product profitability.csv", "/content/Product profitability.csv"],
    "groups": ["/mnt/data/Sales by Groups.csv", "/content/Sales by Groups.csv"],
}

# Select the first existing path from a list (supports Colab + local)
def pick_path(cands):
    for p in cands:
        if os.path.exists(p):
            return p
    raise FileNotFoundError(f"File not found in: {cands}")

MONTHLY_PATH  = pick_path(CANDIDATE_PATHS["monthly"])
CATEGORY_PATH = pick_path(CANDIDATE_PATHS["category"])
PROFIT_PATH   = pick_path(CANDIDATE_PATHS["profit"])
GROUPS_PATH   = pick_path(CANDIDATE_PATHS["groups"])

print("Using:")
print(" Monthly:", MONTHLY_PATH)
print(" Category:", CATEGORY_PATH)
print(" Profit:", PROFIT_PATH)
print(" Groups:", GROUPS_PATH)

# ---------- Helpers ----------
def to_num(s):
    """Convert messy numeric strings to float: removes commas/extra spaces; returns NaN if not numeric."""
    return pd.to_numeric(
        pd.Series(s, dtype="object").astype(str)
          .str.replace(",", "", regex=False)
          .str.replace("\u00a0", " ", regex=False)
          .str.strip()
          .replace({"": np.nan, "nan": np.nan, "None": np.nan}),
        errors="coerce"
    )

def norm_branch(x):
    """Standardize branch naming (spacing + common prefixes/variants)."""
    x = str(x).replace("\u00a0", " ")
    x = re.sub(r"\s+", " ", x).strip()
    x = re.sub(r"(?i)^Branch:\s*", "", x)               # drop "Branch:" prefix
    x = re.sub(r"(?i)^Stories\s*-\s*", "Stories ", x)   # "Stories - X" -> "Stories X"
    x = x.replace("Stories.", "Stories.")
    x = re.sub(r"\s+", " ", x).strip()
    return x

def looks_like_branch(s):
    """Heuristic: treat any text containing 'Stories' as a branch row/header."""
    s = str(s)
    return bool(re.search(r"\bStories\b", s, flags=re.I))

def drop_report_noise(df, text_col=0):
    """Remove repeated report header/footer lines (Page/Printed/Report) and fully empty rows."""
    col = df[text_col].astype(str)
    noise = (
        col.str.contains(r"Page\s+\d+\s+of\s+\d+", case=False, na=False) |
        col.str.contains(r"Printed", case=False, na=False) |
        col.str.contains(r"Report", case=False, na=False)
    )
    df = df[~noise].copy()
    df = df[~df.apply(lambda r: r.astype(str).str.strip().eq("").all(), axis=1)].copy()
    return df

def is_date_stamp(s):
    """Detect date-stamp artifacts like '22-Jan-26' that sometimes appear as rows."""
    return bool(re.match(r"^\d{1,2}-[A-Za-z]{3}-\d{2}$", str(s).strip()))

# ==========================================================
# 1) MONTHLY SALES CLEAN (wide)
# Goal: parse raw report blocks (Jan–Sep repeated sections + Oct–Dec block) into one table
# ==========================================================
def clean_monthly_sales(path):
    raw = pd.read_csv(path, header=None, encoding="utf-8-sig")
    raw = drop_report_noise(raw, text_col=0)

    months_full = ["January","February","March","April","May","June","July","August","September"]
    month_short = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep"]
    full_to_short = dict(zip(months_full, month_short))

    # Locate header rows that contain month names (pagination repeats these headers)
    jan_headers = raw[raw.apply(lambda r: r.astype(str).str.contains(r'\bJanuary\b', na=False).any(), axis=1)].index.tolist()
    oct_headers = raw[raw.apply(lambda r: r.astype(str).str.contains(r'\bOctober\b', na=False).any(), axis=1)].index.tolist()
    if not jan_headers or not oct_headers:
        raise ValueError("Monthly sales: could not find January/October header blocks.")

    first_oct = oct_headers[0]

    # ---- Jan–Sep sections
    sections = []
    for i, h in enumerate(jan_headers):
        start = h + 1
        end = jan_headers[i+1] if i+1 < len(jan_headers) else first_oct
        if start >= first_oct:
            break

        header = raw.loc[h]

        # Map each month to the correct column index by scanning the header row
        month_cols = {}
        for col in raw.columns:
            cell = str(header[col])
            for m in months_full:
                if re.search(rf"\b{m}\b", cell, flags=re.IGNORECASE):
                    month_cols[full_to_short[m]] = col

        # Skip blocks where January column couldn't be detected
        if "Jan" not in month_cols:
            continue

        block = raw.iloc[start:end].copy()

        # Keep only actual branch rows
        block = block[block[1].astype(str).apply(looks_like_branch)].copy()

        # Forward-fill year values inside the block
        block[0] = block[0].replace("", np.nan).ffill()

        # Extract Year, Branch, and whichever month columns exist in this block
        cols = [0, 1] + [month_cols[m] for m in month_short if m in month_cols]
        block = block[cols].copy()

        # Rename columns into stable names (Year, Branch, Jan..Sep)
        rename_map = {0: "Year", 1: "Branch"}
        for m in month_short:
            if m in month_cols:
                rename_map[month_cols[m]] = m
        block = block.rename(columns=rename_map)

        # Convert month values to numeric
        for m in month_short:
            if m in block.columns:
                block[m] = to_num(block[m])

        block["Year"] = pd.to_numeric(block["Year"], errors="coerce").ffill()
        block["Branch"] = block["Branch"].apply(norm_branch)

        sections.append(block)

    df_a = pd.concat(sections, ignore_index=True)
    df_a = df_a.dropna(subset=["Year","Branch"])
    df_a["Year"] = df_a["Year"].astype(int)

    # If repeated pages create duplicates, keep the last occurrence
    df_a = df_a.drop_duplicates(["Year","Branch"], keep="last")

    # ---- Oct–Dec block
    start_b = first_oct + 1
    df_b_raw = raw.iloc[start_b:].copy()

    df_b = df_b_raw[df_b_raw[1].astype(str).apply(looks_like_branch)].copy()
    df_b[0] = df_b[0].replace("", np.nan).ffill()

    df_b = df_b[[0,1,2,3,4,5]].copy()
    df_b.columns = ["Year","Branch","Oct","Nov","Dec","Total"]

    for c in ["Oct","Nov","Dec","Total"]:
        df_b[c] = to_num(df_b[c])

    df_b["Year"] = pd.to_numeric(df_b["Year"], errors="coerce").ffill().astype(int)
    df_b["Branch"] = df_b["Branch"].apply(norm_branch)

    # If repeated pages create duplicates, keep the last occurrence
    df_b = df_b.drop_duplicates(["Year","Branch"], keep="last")

    # Merge Jan–Sep with Oct–Dec
    monthly_wide = df_a.merge(df_b, on=["Year","Branch"], how="left")

    # Ensure all expected columns exist even if missing from the raw file
    for m in ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec","Total"]:
        if m not in monthly_wide.columns:
            monthly_wide[m] = np.nan

    monthly_wide = monthly_wide[["Year","Branch","Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec","Total"]]
    monthly_wide = monthly_wide.sort_values(["Year","Branch"]).reset_index(drop=True)
    return monthly_wide

monthly_wide = clean_monthly_sales(MONTHLY_PATH)
print("monthly_wide:", monthly_wide.shape)
display(monthly_wide.head(10))

# ==========================================================
# 2) CATEGORY SUMMARY CLEAN (Branch + Category rows)
# Goal: keep only BEVERAGES/FOOD/TOTAL BY BRANCH rows and compute a safe Revenue
# Note: Revenue bug is handled by using Revenue = Total_Cost + Total_Profit
# ==========================================================
def clean_category_summary(path):
    raw = pd.read_csv(path, header=None, encoding="utf-8-sig")
    raw = drop_report_noise(raw, text_col=0)

    # Find likely branch column among first few cols (contains "Stories")
    branch_col = None
    for c in raw.columns[:3]:
        if raw[c].astype(str).str.contains(r"\bStories\b", case=False, na=False).any():
            branch_col = c
            break
    if branch_col is None:
        branch_col = 0

    # Find likely category column among first few cols (BEVERAGES/FOOD/TOTAL BY BRANCH)
    cat_col = None
    for c in raw.columns[:3]:
        if raw[c].astype(str).str.contains(r"^(BEVERAGES|FOOD|Total By Branch:)$", case=False, na=False).any():
            cat_col = c
            break
    if cat_col is None:
        cat_col = branch_col

    # Build Branch values via forward fill
    raw["Branch"] = np.where(raw[branch_col].astype(str).apply(looks_like_branch), raw[branch_col].astype(str), np.nan)
    raw["Branch"] = raw["Branch"].ffill().apply(norm_branch)

    # Keep only category summary rows
    cat_vals = raw[cat_col].astype(str).str.strip()
    keep_mask = cat_vals.str.match(r"^(BEVERAGES|FOOD|Total By Branch:)$", case=False, na=False)
    df = raw[keep_mask].copy()
    df["Category"] = cat_vals[keep_mask].str.upper()

    # Choose numeric columns automatically by completeness
    exclude = {branch_col, cat_col}
    num_candidates = [c for c in raw.columns if c not in exclude and c != "Branch"]

    cand_scores = []
    for c in num_candidates:
        nn = to_num(df[c]).notna().mean()
        cand_scores.append((c, nn))
    cand_scores = sorted(cand_scores, key=lambda x: x[1], reverse=True)
    best_cols = [c for c,score in cand_scores if score > 0.3][:6]
    if len(best_cols) < 6:
        best_cols = num_candidates[:6]

    df_out = pd.DataFrame({
        "Branch": df["Branch"],
        "Category": df["Category"],
        "Qty": to_num(df[best_cols[0]]),
        "Total_Price_raw": to_num(df[best_cols[1]]),
        "Total_Cost": to_num(df[best_cols[2]]),
        "Cost_Percent": to_num(df[best_cols[3]]),
        "Total_Profit": to_num(df[best_cols[4]]),
        "Profit_Percent": to_num(df[best_cols[5]]),
    })

    # Safe revenue computation (avoids POS display truncation issues)
    df_out["Revenue"] = df_out["Total_Cost"] + df_out["Total_Profit"]
    df_out["Margin"] = np.where(df_out["Revenue"] > 0, df_out["Total_Profit"] / df_out["Revenue"], np.nan)

    df_out = df_out.sort_values(["Branch","Category"]).reset_index(drop=True)
    return df_out

category_branch = clean_category_summary(CATEGORY_PATH)
print("category_branch:", category_branch.shape)
display(category_branch.head(12))

# ==========================================================
# 3) PRODUCT PROFITABILITY CLEAN (preserve hierarchy)
# Goal: reconstruct Branch/Service/Category/Section context, then keep product rows
# Also removes:
# - fake Branch rows ("Stories" / "Stories.")
# - date-stamp artifacts (e.g., "22-Jan-26")
# ==========================================================
def clean_product_profitability(path):
    raw = pd.read_csv(path, header=None, encoding="utf-8-sig")
    raw = drop_report_noise(raw, text_col=0)

    text_col = 0
    txt = raw[text_col].astype(str).str.replace("\u00a0", " ").str.strip()

    raw["Branch"] = np.where(txt.apply(looks_like_branch), txt, np.nan)
    raw["Branch"] = raw["Branch"].ffill().apply(norm_branch)

    raw["Service"] = np.where(txt.str.match(r"^(TAKE\s*AWAY|TABLE)$", case=False, na=False), txt.str.upper(), np.nan)
    raw["Service"] = raw["Service"].ffill()

    raw["Category"] = np.where(txt.str.match(r"^(BEVERAGES|FOOD)$", case=False, na=False), txt.str.upper(), np.nan)
    raw["Category"] = raw["Category"].ffill()

    raw["Section"] = np.where(txt.str.contains(r"SECTION", case=False, na=False), txt.str.upper(), np.nan)
    raw["Section"] = raw["Section"].ffill()

    # Keep only rows that contain at least one numeric value (actual data rows)
    numeric = raw.apply(pd.to_numeric, errors="coerce")
    has_numbers = numeric.notna().any(axis=1)

    df = raw[has_numbers].copy()
    prod = df[text_col].astype(str).str.strip()

    # Remove totals that leak into the product column
    bad = (
        prod.str.contains(r"^Total", case=False, na=False) |
        prod.str.contains(r"^Grand", case=False, na=False)
    )
    df = df[~bad].copy()
    df["Product"] = prod

    # Score numeric columns and pick the best candidates (kept for robustness)
    num_cols = [c for c in raw.columns if c not in ["Branch","Service","Category","Section","Product"]]
    scores = []
    for c in num_cols:
        nn = to_num(df[c]).notna().mean()
        scores.append((c, nn))
    scores = sorted(scores, key=lambda x: x[1], reverse=True)
    best = [c for c,score in scores if score > 0.3][:6]
    if len(best) < 6:
        best = num_cols[:6]

    out = pd.DataFrame({
    "Branch": df["Branch"],
    "Service": df["Service"],
    "Category": df["Category"],
    "Section": df["Section"],
    "Product": df["Product"],
    "Qty":             to_num(df[1]),
    "Total_Price_raw": to_num(df[2]),
    "Total_Cost":      to_num(df[4]),
    "Cost_Percent":    to_num(df[5]),
    "Total_Profit":    to_num(df[6]),
    "Profit_Percent":  to_num(df[8]),
})

    # Revenue: prefer reported Total_Price when valid; otherwise fallback to Cost + Profit
    out["Revenue"] = np.where(out["Total_Price_raw"] > 0, out["Total_Price_raw"], out["Total_Cost"] + out["Total_Profit"])
    out["Margin"] = np.where(out["Revenue"] > 0, out["Total_Profit"] / out["Revenue"], np.nan)

    out = out.dropna(subset=["Branch","Product"]).reset_index(drop=True)

    # Remove fake branch rows like "Stories" or "Stories."
    out = out[~out["Branch"].astype(str).str.fullmatch(r"Stories\.?", case=False, na=False)].copy()

    # Remove date-stamp artifacts that appear as products
    out = out[~out["Product"].astype(str).apply(is_date_stamp)].copy()

    out.reset_index(drop=True, inplace=True)
    return out

product_profit = clean_product_profitability(PROFIT_PATH)
print("product_profit:", product_profit.shape)
display(product_profit.head(15))

# ==========================================================
# 4) SALES BY GROUPS CLEAN (FIXED)
# Goal: reconstruct Branch/Division/Group context, keep product-level numeric rows,
# and ensure Group is filled correctly within each Branch+Division.
# ==========================================================
def clean_sales_by_groups(path):
    raw = pd.read_csv(path, header=None, encoding="utf-8-sig")
    raw = drop_report_noise(raw, text_col=0)

    text_col = 0
    txt = raw[text_col].astype(str).str.replace("\u00a0", " ").str.strip()

    # ---- Branch
    raw["Branch"] = np.where(txt.apply(looks_like_branch), txt, np.nan)
    raw["Branch"] = raw["Branch"].ffill().apply(norm_branch)

    # ---- Division (supports both "Division:" and "Total by Division:" rows)
    raw["Division"] = np.where(
        txt.str.match(r"(?i)^(Division:|Total\s+by\s+Division:)\s*", na=False),
        txt.str.replace(r"(?i)^(Division:|Total\s+by\s+Division:)\s*", "", regex=True).str.strip(),
        np.nan
    )
    raw["Division"] = raw["Division"].ffill()

    # ---- Group (supports both "Group:" and "Total by Group:" rows)
    raw["Group"] = np.where(
        txt.str.match(r"(?i)^(Group:|Total\s+by\s+Group:)\s*", na=False),
        txt.str.replace(r"(?i)^(Group:|Total\s+by\s+Group:)\s*", "", regex=True).str.upper().str.strip(),
        np.nan
    )
    raw["Group"] = raw["Group"].ffill()

    # ---- Keep only rows that contain at least one numeric value (actual data rows)
    numeric = raw.apply(pd.to_numeric, errors="coerce")
    has_numbers = numeric.notna().any(axis=1)
    df = raw[has_numbers].copy()

    prod = df[text_col].astype(str).str.strip()
    df["Product"] = prod

    # ---- Remove totals and headers that may appear in the Product field
    bad = (
        prod.str.match(r"(?i)^(Total\s+by\s+Group|Total\s+by\s+Branch|Total\s+by\s+Division|Grand\s+Total)\s*:?", na=False) |
        prod.str.match(r"(?i)^(Group:|Division:)", na=False)
    )
    df = df[~bad].copy()

    # ---- Detect numeric columns to map into Quantity and Amount
    num_cols = [c for c in raw.columns if c not in ["Branch","Division","Group","Product"]]

    scores = []
    for c in num_cols:
        nn = to_num(df[c]).notna().mean()
        scores.append((c, nn))
    scores = sorted(scores, key=lambda x: x[1], reverse=True)

    best = [c for c,score in scores if score > 0.3][:2]
    if len(best) < 2:
        best = num_cols[:2]

    out = pd.DataFrame({
        "Branch": df["Branch"],
        "Division": df["Division"],
        "Group": df["Group"],
        "Product": df["Product"],
        "Quantity": to_num(df[best[0]]),
        "Amount": to_num(df[best[1]]),
    }).dropna(subset=["Branch","Product"]).reset_index(drop=True)

    # ---- Remove fake branch rows like "Stories" or "Stories."
    out = out[~out["Branch"].astype(str).str.fullmatch(r"Stories\.?", case=False, na=False)].copy()

    # ---- Remove date-stamp artifacts that appear as products
    out = out[~out["Product"].astype(str).apply(is_date_stamp)].copy()

    # ---- Ensure Group is filled within each Branch+Division (handles gaps from report structure)
    out["Group"] = out.groupby(["Branch","Division"])["Group"].transform(lambda s: s.ffill().bfill())

    out.reset_index(drop=True, inplace=True)
    return out


groups_sales = clean_sales_by_groups(GROUPS_PATH)
print("groups_sales:", groups_sales.shape)
display(groups_sales.head(15))

# ==========================================================
# Quick sanity checks
# Purpose: confirm the cleaned tables have the expected coverage and structure
# ==========================================================
print("\nSanity checks:")
print("Monthly branches:", monthly_wide["Branch"].nunique(), "Years:", sorted(monthly_wide["Year"].unique().tolist()))
print("Category branches:", category_branch["Branch"].nunique(), "Categories:", category_branch["Category"].unique().tolist())
print("Product branches:", product_profit["Branch"].nunique(), "Services:", product_profit["Service"].dropna().unique()[:5])
print("Groups branches:", groups_sales["Branch"].nunique(), "Example groups:", groups_sales["Group"].dropna().unique()[:5])

# Optional: save cleaned outputs
# monthly_wide.to_csv("clean_monthly_wide.csv", index=False)
# category_branch.to_csv("clean_category_branch.csv", index=False)
# product_profit.to_csv("clean_product_profitability.csv", index=False)
# groups_sales.to_csv("clean_groups_sales.csv", index=False)
```

**PRODUCT CLUSTERING AND LOCAION ANALYSIS**

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# ============================================================
# PART 1 — PRODUCT CLUSTERING (3 CLUSTERS)
# Goal:
# - Aggregate product performance across all branches
# - Cluster products using (Volume, Profit, Margin)
# - Convert clusters into business labels: Star / Core Products / Low Performer
# ============================================================

# Build product-level table: total quantity, total profit, and average margin
df_products = product_profit.groupby('Product').agg(
    Total_Qty=('Qty','sum'),
    Total_Profit=('Total_Profit','sum'),
    Avg_Margin=('Margin','mean')
).dropna().reset_index()

# Remove products with tiny sales volume to reduce noise and unstable margins
df_products = df_products[df_products['Total_Qty'] > 10]

# Standardize features so clustering is not dominated by large numeric scales
scaler = StandardScaler()
X = scaler.fit_transform(df_products[['Total_Qty','Total_Profit','Avg_Margin']])

# Train KMeans with 3 clusters to separate products into strategic groups
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df_products['Cluster'] = kmeans.fit_predict(X)

# ============================================================
# CLUSTER SCORING (to assign business meaning)
# Goal:
# - Compute average (Qty, Profit, Margin) per cluster
# - Create a weighted score to rank clusters from best to worst
# - Map clusters into readable labels
# ============================================================

cluster_summary = df_products.groupby('Cluster')[['Total_Qty','Total_Profit','Avg_Margin']].mean()

# Compute a composite score:
# - Profit weight: 40%
# - Margin weight: 35%
# - Volume weight: 25%
cluster_summary['Score'] = (
    (cluster_summary['Total_Profit'] / cluster_summary['Total_Profit'].max() * 0.4) +
    (cluster_summary['Avg_Margin'] / cluster_summary['Avg_Margin'].max() * 0.35) +
    (cluster_summary['Total_Qty'] / cluster_summary['Total_Qty'].max() * 0.25)
)

# Rank clusters by score and assign the best to "Star", etc.
sorted_clusters = cluster_summary['Score'].sort_values(ascending=False).index

label_map = {
    sorted_clusters[0]: 'Star',
    sorted_clusters[1]: 'Core Products',
    sorted_clusters[2]: 'Low Performer'
}

df_products['Label'] = df_products['Cluster'].map(label_map)

# ============================================================
# VISUALIZATION
# Goal:
# - Show products on a strategic map:
#   X = total quantity (log scale), Y = average margin
# - Color by assigned label for interpretation
# ============================================================

plt.figure(figsize=(10,6))

colors = {
    'Star':'gold',
    'Core Products':'steelblue',
    'Low Performer':'red'
}

for label, group in df_products.groupby('Label'):
    plt.scatter(
        group['Total_Qty'],
        group['Avg_Margin'].clip(0,1),
        label=label,
        color=colors[label],
        alpha=0.7,
        s=70,
        edgecolors='w'
    )

plt.xscale('log')
plt.xlabel('Total Quantity Sold (Log Scale)')
plt.ylabel('Average Margin')
plt.title('Product Strategic Mapping (3 Clusters)')
plt.legend()
plt.grid(True, which="both", alpha=0.2)
plt.tight_layout()
plt.savefig('product_clusters_3.png', dpi=150)
plt.show()

# ============================================================
# PRINT TOP PRODUCTS PER CLUSTER
# Goal:
# - For each business label, list the top 5 products by total profit
# - Useful for reporting and recommendations
# ============================================================

for label in ['Star','Core Products','Low Performer']:
    top = df_products[df_products['Label']==label] \
        .nlargest(5,'Total_Profit')[['Product','Total_Qty','Total_Profit','Avg_Margin']]

    print(f"\n--- {label} ---")
    print(top.to_string(index=False))

# ============================================================
# PART 2 — BRANCH RANKING
# Goal:
# - Use branch-level totals to create a performance score
# - Score combines Profit (50%), Revenue (30%), Margin (20%)
# - Split branches into tiers: Top / Mid / Underperformer
# ============================================================

# Keep only TOTAL BY BRANCH rows and select needed columns
df_branch = category_branch[category_branch['Category'] == 'TOTAL BY BRANCH:'].copy()
df_branch = df_branch[['Branch','Qty','Revenue','Total_Profit','Margin']].dropna()

# Normalize features to 0–1 so weighting is meaningful
scaler2 = MinMaxScaler()
cols = ['Revenue','Total_Profit','Margin','Qty']
df_branch_norm = pd.DataFrame(scaler2.fit_transform(df_branch[cols]), columns=cols, index=df_branch.index)

# Weighted score emphasizing profit most
df_branch['Score'] = (
    df_branch_norm['Total_Profit'] * 0.5 +
    df_branch_norm['Revenue']      * 0.3 +
    df_branch_norm['Margin']       * 0.2
)

# Rank branches and assign tiers based on rank bins
df_branch = df_branch.sort_values('Score', ascending=False).reset_index(drop=True)
df_branch['Rank'] = df_branch.index + 1
df_branch['Tier'] = pd.cut(df_branch['Rank'], bins=[0,8,17,30], labels=['Top','Mid','Underperformer'])

print("\nBranch Rankings:")
print(df_branch[['Rank','Branch','Score','Tier']].to_string(index=False))

# Plot branch score leaderboard (horizontal bar chart)
plt.figure(figsize=(10,8))
bar_colors = df_branch['Tier'].map({'Top':'teal','Mid':'orange','Underperformer':'red'})
plt.barh(df_branch['Branch'], df_branch['Score'], color=bar_colors)
plt.xlabel('Performance Score')
plt.title('Branch Ranking')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig('branch_ranking.png', dpi=150)
plt.show()

# ============================================================
# PART 3 — LOCATION ANALYSIS
# Goal:
# - Compare "Street" branches vs "Mall/Special" branches
# - Show average revenue, average profit, and average margin by location type
# ============================================================

# Define mall/special branches list (everything else is treated as street)
mall_branches = ['Stories Centro Mall', 'Stories Le Mall', 'Stories LAU',
                 'Stories Airport', 'Stories Event Starco']

df_branch_loc = category_branch[category_branch['Category'] == 'TOTAL BY BRANCH:'].copy()
df_branch_loc['Location_Type'] = df_branch_loc['Branch'].apply(
    lambda x: 'Mall/Special' if x in mall_branches else 'Street'
)

# Aggregate performance by location type
loc_summary = df_branch_loc.groupby('Location_Type').agg(
    Avg_Revenue=('Revenue','mean'),
    Avg_Profit=('Total_Profit','mean'),
    Avg_Margin=('Margin','mean'),
    Branch_Count=('Branch','count')
).reset_index()

print("\nLocation Analysis:")
print(loc_summary.to_string(index=False))

# Plot 3 side-by-side bar charts (revenue / profit / margin)
fig, axes = plt.subplots(1, 3, figsize=(14,5))
axes[0].bar(loc_summary['Location_Type'], loc_summary['Avg_Revenue'], color=['steelblue','orange'])
axes[0].set_title('Avg Revenue')
axes[1].bar(loc_summary['Location_Type'], loc_summary['Avg_Profit'], color=['steelblue','orange'])
axes[1].set_title('Avg Profit')
axes[2].bar(loc_summary['Location_Type'], loc_summary['Avg_Margin'], color=['steelblue','orange'])
axes[2].set_title('Avg Margin')
plt.suptitle('Mall vs Street Performance', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('location_analysis.png', dpi=150)
plt.show()

# Quick reporting prints for the final summary section
print("\n--- Product Stars ---")
print(df_products[df_products['Label']=='Star'].nlargest(5,'Total_Profit')[['Product','Total_Qty','Total_Profit']].to_string(index=False))
print("\n--- Top 5 Branches ---")
print(df_branch[['Rank','Branch','Score','Tier']].head(5).to_string(index=False))
```

**INTERACTIVE DASHBOARD**

```python
# ============================================================
# FULL COLAB INTERACTIVE CEO DASHBOARD + DECISION SUPPORT
#
# REQUIRES:
#   df_wide    (Year, Branch, Jan..Dec)
#   df_product (Product, Qty, Total_Profit)  [ideally Revenue OR Total_Cost too]
#
# OPTIONAL:
#   df_pred_2026 (Branch, Date, Predicted_Sales)
#
# NOTES:
# - Menu Engineering uses Profit per Unit = Total_Profit / Qty
# - Option B impact estimates are "what-if" approximations (explainable, not accounting-level)
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output

MONTHS = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
MONTH_MAP = {m:i+1 for i,m in enumerate(MONTHS)}

# -----------------------------
# Convert df_wide to long format
# -----------------------------
def to_long_monthly(df):
    df2 = df.copy()
    for m in MONTHS:
        if m in df2.columns:
            df2[m] = pd.to_numeric(df2[m], errors="coerce")

    long = df2.melt(
        id_vars=["Year","Branch"],
        value_vars=[m for m in MONTHS if m in df2.columns],
        var_name="Month",
        value_name="Sales"
    )

    long["Sales"] = pd.to_numeric(long["Sales"], errors="coerce").fillna(0)
    long["MonthNum"] = long["Month"].map(MONTH_MAP)
    long["Date"] = pd.to_datetime(dict(year=long["Year"], month=long["MonthNum"], day=1), errors="coerce")
    long["Branch"] = long["Branch"].astype(str).str.strip()
    long["Year"] = pd.to_numeric(long["Year"], errors="coerce")

    long = long.dropna(subset=["Date","Branch","Year"])
    return long.sort_values(["Date","Branch"]).reset_index(drop=True)

# -----------------------------
# Build menu engineering table + quadrant labels
# -----------------------------
def build_menu_engineering_table(df_product):
    needed = {"Product", "Qty", "Total_Profit"}
    if not needed.issubset(df_product.columns):
        return None, "Menu engineering needs columns: Product, Qty, Total_Profit"

    d = df_product.copy()
    d["Product"] = d["Product"].astype(str).str.strip()
    d["Qty"] = pd.to_numeric(d["Qty"], errors="coerce")
    d["Total_Profit"] = pd.to_numeric(d["Total_Profit"], errors="coerce")

    # Optional revenue support if exists
    if "Revenue" in d.columns:
        d["Revenue"] = pd.to_numeric(d["Revenue"], errors="coerce")
    elif {"Total_Cost", "Total_Profit"}.issubset(d.columns):
        d["Total_Cost"] = pd.to_numeric(d["Total_Cost"], errors="coerce")
        d["Revenue"] = d["Total_Cost"] + d["Total_Profit"]
    else:
        d["Revenue"] = np.nan  # impact estimates will fall back to rough method

    d = d.dropna(subset=["Qty","Total_Profit"])
    d = d[d["Qty"] > 0].copy()

    d["Profit_per_unit"] = d["Total_Profit"] / d["Qty"]

    # Median cut lines (classic)
    x_cut = d["Qty"].median()
    y_cut = d["Profit_per_unit"].median()

    def quadrant(r):
        if r["Qty"] >= x_cut and r["Profit_per_unit"] >= y_cut:
            return "Stars"
        if r["Qty"] >= x_cut and r["Profit_per_unit"] < y_cut:
            return "Plowhorses"
        if r["Qty"] < x_cut and r["Profit_per_unit"] >= y_cut:
            return "Puzzles"
        return "Dogs"

    d["Quadrant"] = d.apply(quadrant, axis=1)

    return d, None

def plot_menu_engineering(d):
    colors = {"Stars":"gold","Plowhorses":"steelblue","Puzzles":"purple","Dogs":"red"}

    x_cut = d["Qty"].median()
    y_cut = d["Profit_per_unit"].median()

    plt.figure(figsize=(10,6))
    for q, g in d.groupby("Quadrant"):
        plt.scatter(
            g["Qty"], g["Profit_per_unit"],
            s=60, alpha=0.7, label=q,
            color=colors.get(q, "gray"),
            edgecolors="white"
        )
    plt.axvline(x_cut, linestyle="--")
    plt.axhline(y_cut, linestyle="--")
    plt.title("Menu Engineering — Popularity vs Profitability (4 Quadrants)")
    plt.xlabel("Popularity (Quantity Sold)")
    plt.ylabel("Profitability (Profit per Unit)")
    plt.grid(True, alpha=0.25)
    plt.legend()
    plt.tight_layout()
    plt.show()

# -----------------------------
# Option B — Recommendation engine
# -----------------------------
def decision_recommendations(d, price_increase_pct, promo_lift_pct, drop_dogs_n):
    """
    price_increase_pct: % price increase applied to Stars + Plowhorses revenue (if revenue available)
    promo_lift_pct: % qty lift applied to Puzzles (assume profit/unit stays same)
    drop_dogs_n: number of dog products to consider removing
    """

    # Top lists
    stars = d[d["Quadrant"]=="Stars"].sort_values("Total_Profit", ascending=False)
    puzzles = d[d["Quadrant"]=="Puzzles"].sort_values("Profit_per_unit", ascending=False)
    plow = d[d["Quadrant"]=="Plowhorses"].sort_values("Qty", ascending=False)
    dogs = d[d["Quadrant"]=="Dogs"].sort_values("Total_Profit", ascending=True)

    # Impact estimates (rough)
    # 1) Price increase on Stars + Plowhorses: revenue * pct -> assume profit increases by same delta (simplified)
    inc = price_increase_pct / 100.0
    promo = promo_lift_pct / 100.0

    rev_base = np.nan
    if d["Revenue"].notna().any():
        rev_base = d[d["Quadrant"].isin(["Stars","Plowhorses"])]["Revenue"].sum()
        profit_uplift_price = rev_base * inc
    else:
        # fallback: use profit as proxy (more conservative)
        profit_base = d[d["Quadrant"].isin(["Stars","Plowhorses"])]["Total_Profit"].sum()
        profit_uplift_price = profit_base * inc

    # 2) Promotion on Puzzles: qty lift * profit/unit
    puzzles_base = d[d["Quadrant"]=="Puzzles"].copy()
    profit_uplift_promo = (puzzles_base["Qty"] * promo * puzzles_base["Profit_per_unit"]).sum()

    # 3) Dropping dogs: show items; no numeric "savings" unless cost data exists
    dogs_to_drop = dogs.head(drop_dogs_n)

    rec = {
        "stars_top": stars.head(5)[["Product","Qty","Total_Profit","Profit_per_unit"]],
        "puzzles_top": puzzles.head(5)[["Product","Qty","Total_Profit","Profit_per_unit"]],
        "plow_top": plow.head(5)[["Product","Qty","Total_Profit","Profit_per_unit"]],
        "dogs_drop": dogs_to_drop[["Product","Qty","Total_Profit","Profit_per_unit"]],
        "profit_uplift_price": float(profit_uplift_price),
        "profit_uplift_promo": float(profit_uplift_promo),
        "total_uplift": float(profit_uplift_price + profit_uplift_promo),
    }
    return rec

# -----------------------------
# Checks
# -----------------------------
if "df_wide" not in globals():
    raise ValueError("df_wide not found. Create df_wide first.")
if "df_product" not in globals():
    raise ValueError("df_product not found. Create df_product first.")

df_long = to_long_monthly(df_wide)

# -----------------------------
# Widgets (Option A filters)
# -----------------------------
year_dropdown = widgets.Dropdown(
    options=sorted(df_long["Year"].dropna().unique()),
    description="Year:"
)

branch_dropdown = widgets.Dropdown(
    options=["All"] + sorted(df_long["Branch"].unique()),
    description="Branch:"
)

# -----------------------------
# Option B controls (what-if)
# -----------------------------
price_slider = widgets.IntSlider(
    value=2, min=0, max=10, step=1,
    description="Price +%:", continuous_update=False
)

promo_slider = widgets.IntSlider(
    value=5, min=0, max=30, step=1,
    description="Promo +%:", continuous_update=False
)

drop_dogs_slider = widgets.IntSlider(
    value=5, min=0, max=20, step=1,
    description="Drop Dogs:", continuous_update=False
)

output = widgets.Output()

# -----------------------------
# Dashboard update
# -----------------------------
def update_dashboard(change=None):
    with output:
        clear_output(wait=True)

        year_sel = year_dropdown.value
        branch_sel = branch_dropdown.value

        # ========= OPTION A (CEO DASHBOARD) =========
        df_f = df_long[df_long["Year"] == year_sel].copy()
        if branch_sel != "All":
            df_f = df_f[df_f["Branch"] == branch_sel]

        monthly = df_f.groupby("Date")["Sales"].sum().sort_index()

        total_sales = df_f["Sales"].sum()
        avg_month = monthly.mean() if len(monthly) else 0
        best_month = monthly.idxmax() if len(monthly) else None
        best_val = monthly.max() if len(monthly) else 0

        print("===== STORIES — EXECUTIVE DASHBOARD (Option A) =====")
        print(f"Year: {int(year_sel)} | Branch: {branch_sel}")
        print(f"Total Sales: {total_sales:,.0f}")
        print(f"Avg Monthly Sales: {avg_month:,.0f}")
        if best_month is not None:
            print(f"Best Month: {best_month.strftime('%b %Y')} ({best_val:,.0f})")
        print("\n")

        # Monthly trend
        plt.figure(figsize=(10,4))
        plt.plot(monthly.index, monthly.values, marker="o")
        plt.title("Monthly Sales Trend")
        plt.xlabel("Month")
        plt.ylabel("Sales")
        plt.grid(True)
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

        # Branch leaderboard
        print("Top Branches (Year Total):")
        leaderboard = (df_long[df_long["Year"] == year_sel]
                       .groupby("Branch")["Sales"]
                       .sum()
                       .sort_values(ascending=False)
                       .head(10))
        display(leaderboard.to_frame("Total Sales"))

        # Product top profit
        if "Total_Profit" in df_product.columns:
            print("\nTop Products by Total Profit:")
            cols_show = [c for c in ["Product","Qty","Revenue","Total_Profit"] if c in df_product.columns]
            # If Revenue isn't in df_product, it won't show; that's ok
            top_products = df_product.sort_values("Total_Profit", ascending=False).head(10)
            display(top_products[cols_show] if cols_show else top_products.head(10))

        # Menu engineering plot
        d_me, err = build_menu_engineering_table(df_product)
        if err:
            print("\n" + err)
            return

        print("\nMenu Engineering (Classic 4 Quadrants):")
        plot_menu_engineering(d_me)

        # Optional forecast chart (company)
        if "df_pred_2026" in globals() and isinstance(df_pred_2026, pd.DataFrame) and not df_pred_2026.empty:
            df_fc = df_pred_2026.copy()
            df_fc["Date"] = pd.to_datetime(df_fc["Date"], errors="coerce")
            df_fc["Predicted_Sales"] = pd.to_numeric(df_fc["Predicted_Sales"], errors="coerce")
            df_fc = df_fc.dropna(subset=["Date","Predicted_Sales"])

            company_fcst = df_fc.groupby("Date")["Predicted_Sales"].sum().sort_index()

            print("\nCompany Forecast (Sum of Branches):")
            plt.figure(figsize=(10,4))
            plt.plot(company_fcst.index, company_fcst.values, marker="o", linestyle="--")
            plt.title("Company Forecast (Feb–Dec 2026)")
            plt.xlabel("Month")
            plt.ylabel("Predicted Sales")
            plt.grid(True)
            plt.xticks(rotation=45)
            plt.tight_layout()
            plt.show()

        # ========= OPTION B (DECISION SUPPORT) =========
        print("\n\n===== DECISION SUPPORT (Option B) — Recommendations =====")

        price_increase_pct = price_slider.value
        promo_lift_pct = promo_slider.value
        drop_dogs_n = drop_dogs_slider.value

        rec = decision_recommendations(
            d_me,
            price_increase_pct=price_increase_pct,
            promo_lift_pct=promo_lift_pct,
            drop_dogs_n=drop_dogs_n
        )

        print(f"\nWhat-if settings:")
        print(f"- Price increase on Stars + Plowhorses: +{price_increase_pct}%")
        print(f"- Promotion lift on Puzzles (volume): +{promo_lift_pct}%")
        print(f"- Consider removing bottom Dogs: {drop_dogs_n} items")

        print("\nRecommended Actions:")
        print("1) Protect & feature these STAR products (high volume, high profit):")
        display(rec["stars_top"])

        print("\n2) Push these PUZZLES (high profit/unit but low sales) with promos / bundles:")
        display(rec["puzzles_top"])

        print("\n3) Review these PLOWHORSES (high sales but low profit/unit) — reprice or reduce cost:")
        display(rec["plow_top"])

        print("\n4) Consider removing / replacing these DOGS (weak volume + weak profit):")
        display(rec["dogs_drop"])

        print("\nEstimated Impact (rough, explainable):")
        print(f"- Profit uplift from pricing: {rec['profit_uplift_price']:,.0f}")
        print(f"- Profit uplift from promos:  {rec['profit_uplift_promo']:,.0f}")
        print(f"= Total estimated uplift:    {rec['total_uplift']:,.0f}")

# Bind
year_dropdown.observe(update_dashboard, names="value")
branch_dropdown.observe(update_dashboard, names="value")
price_slider.observe(update_dashboard, names="value")
promo_slider.observe(update_dashboard, names="value")
drop_dogs_slider.observe(update_dashboard, names="value")

# Display UI
controls_a = widgets.HBox([year_dropdown, branch_dropdown])
controls_b = widgets.HBox([price_slider, promo_slider, drop_dogs_slider])

display(controls_a)
display(controls_b)
display(output)

update_dashboard()
```

**Interactive CEO Dashboard and Decision-Support Tool**

An interactive dashboard was built to translate the analysis into something a decision-maker can use directly. It allows the user to select a year and optionally filter down to a single branch, then instantly generates a high-level performance view: total sales, average monthly sales, the best month of the year, and a clear monthly trend chart. It also produces a yearly leaderboard of the top branches and a ranked list of the most profitable products, so performance can be assessed quickly without manually recalculating totals every time.

On the product side, the tool adds a “menu engineering” layer by measuring both popularity and profitability. Popularity is captured by quantity sold, while profitability is captured by profit per unit (total profit divided by total quantity). Using these two dimensions, each product is automatically classified into one of four strategic groups: high-volume/high-profit items to protect and feature, high-volume/low-profit items to optimize, low-volume/high-profit items to promote, and low-volume/low-profit items to reconsider. This is visualized in a scatter chart that makes it easy to spot which products deserve pricing attention, which should be bundled or promoted, and which may be adding complexity without meaningful financial return.

Finally, the dashboard includes a simple, explainable “what-if” decision layer. The user can simulate small pricing increases on high-impact items, simulate promotional lifts on under-sold but high-profit items, and choose how many weak products to consider removing. Based on these settings, the tool outputs recommended action lists (what to protect, what to promote, what to reprice, what to drop) and provides rough profit uplift estimates. The intent is not to replace accounting, but to provide a fast planning tool that helps prioritize actions and quantify the direction and scale of potential impact.
